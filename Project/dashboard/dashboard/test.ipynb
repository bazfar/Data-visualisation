```python
pov = pd.read_excel('app/geo/Data/poverty_rate.xlsx', decimal=',')
```

```python
pov.columns
```

```python
pov = pov.drop(columns=['Unnamed: 1']).dropna()
pov = pov.rename(columns={'Taux de pauvreté': 'poverty_rate', 'Gouvernorat:': 'gouvernorat'})
pov['gouvernorat'] = pov['gouvernorat'].apply(lambda x: x.title())
```

```python
pov.loc[len(pov)] = {'gouvernorat': 'Tunisie', 'poverty_rate': pov.poverty_rate.mean()}
```

```python
pov
```

```python
population = pd.read_csv('app/geo/Data/population.csv')
```

```python
pop_pov = pd.merge(
    population,
    pov,
    on='gouvernorat',
)
```

```python
pop_pov.to_csv('app/geo/Data/population_poverty.csv', index=False)
```
